Is posting time a factor of the popularity of a YouTube video? 

```python
import datetime as dt
%pylab inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['publish_hour'] = df['publish_time'].str.extract('\s(\d{2})')
publish_views = df.groupby(by = 'publish_hour')['views'].mean()
publish_views_df = pd.DataFrame(publish_views)
publish_views_df.head(3)
```

```python
axis1 = sns.barplot(x = publish_views_df.index, y = publish_views_df.views, palette=['Orange']);
axis1.set_title('Average Views of Videos within 24 Hours', fontdict={'fontsize': 15}, pad=20)
plt.ylabel('Average Views')
plt.xlabel('Publish Hour')
```

Observation: there is a peak of average views of trending videos published between 4 a.m. and 5 a.m.

Exploring what happens to this time period...

```python
df = df.reset_index()
```

```python
category_hour = df.groupby(by=['category_id', 'publish_hour'])['views'].mean().unstack()
category_hour.fillna(0, inplace=True)
```

```python
table2 = pd.DataFrame(category_hour[['04','05']]).unstack()
table2 = table2.reset_index()
table2 = table2.rename(columns={0:'average views'})
```

```python
axis2 = sns.barplot(x='category_id', y='views', data=table2, hue='publish_hour')
axis2.set_title('Average Views for Different Types of Videos at 4 and 5 a.m.', fontdict={'fontsize': 15}, pad=20)
plt.ylabel('Average Views')
plt.xlabel('Category')
plt.xticks(rotation='vertical');
```

Obeservations: Among all videos, the music video has the highest average views during 4 a.m and 5 a.m., indicating that the category is responsible for the abnormal peak of average views in 24 hours.

Does the music video get the similar high average views per hour?

```python
table1 = pd.DataFrame(category_hour.loc['Music',]).unstack()
table1 = table1.reset_index()
table1 = table1.rename(columns={0:'average_views'})
```

```python
axis3 = sns.barplot(x='publish_hour', y='average_views', data=table1)
axis3.set_title('Average Views for Music Videos within 24 Hours', fontdict={'fontsize': 15}, pad=20)
plt.ylabel('Average Views')
plt.xlabel('Publish Hour');
```

Reference: 
1. For the music videos, the similar high peak of average views also appears between 4 a.m. and 5 a.m., indicating that music videos which are published during this time period always recieve more views. After reviewing the dataset, the abnormal phenomenon is largely due to a channel named ChildishGambinoVEVO which owns the highest average views. 

```python
df['time_interval'] = df['trending_date'].apply(pd.to_datetime) - df['publish_date'].apply(pd.to_datetime)
time_interval_df = pd.DataFrame(df['time_interval'].value_counts().sort_index())
time_interval_df = time_interval_df.rename(columns = {'time_interval':'Num_trending_videos'})
time_interval_df.index = time_interval_df.index.days
time_interval_df['Percentage'] = time_interval_df['Num_trending_videos']/sum(time_interval_df['Num_trending_videos'])
axis = sns.barplot(x = time_interval_df.index, y = time_interval_df['Percentage'], palette=['Green']);
xlim(0, 7)
plt.xlabel('Time Interval')
```

Reference: Longer a video posted on Youtube is, less likely it becomes a trending video.
